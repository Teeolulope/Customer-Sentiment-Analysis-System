```python
Customer_name = 'Grace Okafor'
Product_purchased = 'Wireless Headphones'
Satisfaction_rating = 4
Delivery_rating = 5
Product_quality_rating = 4
Customer_support_rating = 3
Number_of_complaints = 1
Repeat_customer = True
Average_rating = (Satisfaction_rating + Delivery_rating + Product_quality_rating + Customer_support_rating)/4
Number_of_complaints += 1
Highly_satisfied = (Average_rating > 4.5)
High_complaint_level = (Number_of_complaints >= 2)
Immediate_follow_up = (Number_of_complaints >= 2 and not Average_rating >=4.5)
brand_advocate = (Average_rating >= 4.5 and Repeat_customer)
Special_appreciation = (brand_advocate or Average_rating > 4.8)


print ('\n==============================================')
print ('CUSTOMER SENTIMENT ANALYSIS REPORT')
print ('==============================================')      
print ('Customer Name:', Customer_name)
print ('Product Purchased:', Product_purchased)
print ('Average Satisfaction Rating:', Average_rating)
print ('Total Complaints:', Number_of_complaints)
print ('Repeat Customer:', Repeat_customer )
print ('Highly Satisfied:', Highly_satisfied )
print ('High Complaint Level:', High_complaint_level)
print ('Immediate Follow-up Required:', Immediate_follow_up)
print ('Brand Advocate:',brand_advocate)  
print ('Special Appreciation Required:', Special_appreciation)
```
